```python
def plot_signal(label, signal, title):
    plt.plot(signal)
    plt.title(f"{label} for {title}")
    plt.xlabel("time")
    plt.ylabel("amplitude")
    plt.show()
```

```python
def replace_zeros_with_average(signal):
    signal = signal.copy()
    n = len(signal)
    for i in range(656,n):
        if signal[i] == 0:
            # Find the previous non-zero value
            prev_val = None
            for j in range(i - 1, -1, -1):
                if signal[j] != 0:
                    prev_val = signal[j]
                    break
            
            # Find the next non-zero value
            next_val = None
            for k in range(i + 1, n):
                if signal[k] != 0:
                    next_val = signal[k]
                    break
            
            # Replace zero with the average of previous and next non-zero values
            if prev_val is not None and next_val is not None:
                signal[i] = (prev_val + next_val) / 2
            elif prev_val is not None:
                signal[i] = prev_val
            elif next_val is not None:
                signal[i] = next_val

    return signal

spo2_mild_clean = replace_zeros_with_average(spo2_mild)

amplitude_values = spo2_mild_clean[700:800]
print(amplitude_values)

plot_signal("SpO2", spo2_mild_clean, "mild")
```

```python
signal_labels = severe_file.getSignalLabels()
files = [severe_file, normal_file, moderate_file, mild_file]
means_dict = {} 

for label in signal_labels:
    means = []
    for file in files:
        mean = np.mean(file.readSignal(file.getSignalLabels().index(label)))
        means.append(mean)
    means_dict[label] = means

print(means_dict)
categories = ['severe', 'normal', 'moderate', 'mild']

for label in signal_labels:
    fig, ax = plt.subplots(figsize=(4, 3))
    ax.bar(categories, means_dict[label], 0.35, label=label)
    
    # Add some text for labels, title, and custom x-axis tick labels
    ax.set_xlabel('Files')
    ax.set_ylabel('Mean Amplitude')
    ax.set_title(f'Mean Amplitude of Signal: {label}')
    ax.legend()

    # Display the plot
plt.show()
        
```

```python
import statistics as st

modes_dict = {} 

for label in signal_labels:
    modes = []
    for file in files:
        mode = st.mode(file.readSignal(file.getSignalLabels().index(label)))
        modes.append(mode)
    modes_dict[label] = modes

print(modes_dict) 
```

```python
medians_dict = {}

for label in signal_labels:
    medians = []
    for file in files:
        median = st.median(file.readSignal(file.getSignalLabels().index(label)))
        medians.append(median)
    medians_dict[label] = medians

print(medians_dict)

categories = ['severe', 'normal', 'moderate', 'mild']

for label in signal_labels:
    fig, ax = plt.subplots(figsize=(4, 3))
    ax.bar(categories, medians_dict[label], 0.35, label=label)
    
    # Add some text for labels, title, and custom x-axis tick labels
    ax.set_xlabel('Files')
    ax.set_ylabel('Median Amplitude')
    ax.set_title(f'Median Amplitude of Signal: {label}')
    ax.legend()

    # Display the plot
plt.show()
```
