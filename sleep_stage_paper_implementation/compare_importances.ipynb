```python
import re
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the text file
with open("c:/Users/anany/Desktop/OSA/sleep_stage_paper_implementation/binary_features.txt", "r") as file:
    data = file.read()

# Extract feature names and scores using regex
matches = re.findall(r"(\w+): ([0-9.]+)", data)

# Create a DataFrame
df = pd.DataFrame(matches, columns=["Feature", "Importance"])
df["Importance"] = df["Importance"].astype(float)

# Count occurrences of each feature and calculate average importance
feature_stats = df.groupby("Feature").agg(
    Count=("Importance", "count"),
    AvgImportance=("Importance", "mean")
).reset_index()

# Sort by count and average importance
feature_stats = feature_stats.sort_values(by=["Count", "AvgImportance"], ascending=False)
```

```python
for item in feature_stats.itertuples():
    print(f"Feature: {item.Feature}, Count: {item.Count}, Avg Importance: {item.AvgImportance:.4f}")
```

```python
plt.figure(figsize=(12, 6))
plt.bar(feature_stats["Feature"], feature_stats["Count"], color="skyblue")
plt.xticks(rotation=90)
plt.title("Feature Frequency in Top 20")
plt.xlabel("Feature")
plt.ylabel("Frequency")
plt.show()
```

```python
# Sort by average importance in descending order
feature_stats_imp = feature_stats.sort_values(by="AvgImportance", ascending=False)

# Print features in sorted order of average importance
for item in feature_stats_imp.itertuples():
    print(f"Feature: {item.Feature}, Count: {item.Count}, Avg Importance: {item.AvgImportance:.4f}")
```

```python
plt.figure(figsize=(12, 6))
plt.bar(feature_stats_imp["Feature"], feature_stats_imp["AvgImportance"], color="orange")
plt.xticks(rotation=90)
plt.title("Average Importance of Features")
plt.xlabel("Feature")
plt.ylabel("Average Importance")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(feature_stats["Count"], feature_stats["AvgImportance"], color="green")
plt.title("Feature Frequency vs. Average Importance")
plt.xlabel("Frequency")
plt.ylabel("Average Importance")
plt.grid(True)

# Annotate top features
for i, row in feature_stats.iterrows():
    if row["Count"] > 2:  # Annotate features appearing more than twice
        plt.text(row["Count"], row["AvgImportance"], row["Feature"], fontsize=8)

plt.show()
```
