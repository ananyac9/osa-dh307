```python
from intervaltree import Interval, IntervalTree

def find_overlaps(t1_tree, t2_intervals):
    overlaps = []
    for start, end in t2_intervals:
        # Query for overlapping intervals in t1
        overlapping_intervals = t1_tree.overlap(start, end)
        if overlapping_intervals:
            overlaps.append((start, end, list(overlapping_intervals)))
    return overlaps

tree_user_wake = IntervalTree(Interval(start, end) for start, end in user_wake_times)
overlaps_wake = find_overlaps(tree_user_wake, machine_wake_times)
print(overlaps_wake)

for times in overlaps_wake:
    print(f"machine interval {times[0:2]} overlaps with user intervals: {times[2]}")
```

```python
def calculate_overlapping_ranges(t1_tree, t2_intervals):
    exact_overlaps = []
    for start, end in t2_intervals:
        overlapping_intervals = t1_tree.overlap(start, end)
        for interval in overlapping_intervals:
            # Calculate exact overlap
            overlap_start = max(interval.begin, start)
            overlap_end = min(interval.end, end)
            exact_overlaps.append((overlap_start, overlap_end))
    return exact_overlaps

exact_overlaps_wake = calculate_overlapping_ranges(tree_user_wake, machine_wake_times)

print(exact_overlaps_wake)
```

```python
def find_intervals_user_diagnostic(type, root, d_limit):
    time_intervals = []
    time_pair = []
    for stage in root.findall('.//User/Stage'):
        stage_type = stage.get('Type')
        start_time = int(stage.get('Start'))

        if start_time > d_limit:
        # If the previous stage's end time hasn't been set yet or it's not d_limit, adjust it
            if last_stage_end_time is not None and last_stage_end_time < d_limit:
                time_pair.append(d_limit)  # Close the interval at d_limit
                time_intervals.append(time_pair)
            break

        if stage_type == type:
            if len(time_pair) == 0:
                time_pair.append(start_time)
        else:
            if len(time_pair) == 1:
                time_pair.append(start_time)
                time_intervals.append(time_pair)
                time_pair = []

        last_stage_end_time = start_time

    return time_intervals
```
