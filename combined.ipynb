```python
for edf_file in edf_files:
    with pyedflib.EdfReader(edf_file) as f:
        print(f.getSampleFrequency(0))  # Check the sampling rate of the first signal
```

```python
for edf_file in edf_files:
    with pyedflib.EdfReader(edf_file) as f:
        print(f.getNSamples())  # Print the number of samples for each signal
```

```python
combined_signals = []

sig = pyedflib.EdfReader(edf_files[0])
sig1 = sig.readSignal(0)
print(len(sig1))

for edf_file in edf_files:
    signal = pyedflib.EdfReader(edf_file).readSignal(0)
    combined_signals.append(signal)

print(len(combined_signals[0]))

combined_signals = np.concatenate(combined_signals)
print(combined_signals.shape)

print(combined_signals[5030000:5030050])
```

```python
import matplotlib.pyplot as plt

def plot_signal(label, signal):
    plt.plot(signal)
    plt.title(f"{label} signal")
    plt.xlabel("time")
    plt.ylabel("amplitude")
    plt.show()

plot_signal("combined", combined_signals)
```
